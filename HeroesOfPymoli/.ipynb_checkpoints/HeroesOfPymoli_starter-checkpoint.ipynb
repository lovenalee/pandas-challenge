```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
unique_SN = purchase_data.drop_duplicates(["SN"])
players_count = len(unique_SN)
Total_Player = pd.DataFrame({"Total Players": [players_count]})
Total_Player
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count=len(purchase_data["Item Name"].unique())
Average_Price = purchase_data["Price"].mean()
purchase_count = len(purchase_data["Purchase ID"].unique())
total_revenue = purchase_data["Price"].sum()
Purchase_Analysis = pd.DataFrame({"Number of Unique Items": [item_count],
                                 "Average Price": [Average_Price],
                                 "Number of Purchase":[purchase_count],
                                 "Total Revenue":[total_revenue]})
Purchase_Analysis["Average Price"]=Purchase_Analysis["Average Price"].astype(float).map("${:,.2f}".format)
Purchase_Analysis["Total Revenue"]=Purchase_Analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
Purchase_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_group = unique_SN.groupby("Gender")
num_of_each_gender = gender_group["Gender"].count()
gender_counts = pd.DataFrame(num_of_each_gender)
gender_counts = gender_counts.rename(columns={"Gender":"Total Count"})
gender_counts
```

```python
gender_percent = pd.DataFrame(num_of_each_gender/players_count)
gender_percent = gender_percent.rename(columns={"Gender":"Percentage of Players"})
gender_percent["Percentage of Players"]=gender_percent["Percentage of Players"].astype(float).map("{:,.2%}".format)
gender_percent
```

```python
Gender_Demographics = gender_counts.merge(gender_percent, on="Gender")
Gender_Demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

group = purchase_data.groupby("Gender")
Quantity = group["Gender"].count()
Purchase_Count = pd.DataFrame(Quantity)
Purchase_Count = Purchase_Count.rename(columns={"Gender":"Purchase Count"})
Purchase_Count
```

```python
price = group["Price"].sum()
Price_count = pd.DataFrame(price)
Price_count = Price_count.rename(columns={"Price":"Total Purchase Value"})
Price_count["Total Purchase Value"]=Price_count["Total Purchase Value"].astype(float).map("${:,.2f}".format)
Price_count
```

```python
average = price/Quantity
Avg_Purchase_P = pd.DataFrame(average)
Avg_Purchase_P = Avg_Purchase_P.rename(columns={0:"Average Purchase Price"})
Avg_Purchase_P["Average Purchase Price"]=Avg_Purchase_P["Average Purchase Price"].astype(float).map("${:,.2f}".format)
Avg_Purchase_P
```

```python

average_pp = price/num_of_each_gender
ATP_PP = pd.DataFrame(average_pp)
ATP_PP = ATP_PP.rename(columns={0:"Avg Total Purchase per Person"})
ATP_PP["Avg Total Purchase per Person"]=ATP_PP["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
ATP_PP
```

```python
DF1 = Purchase_Count.merge(Avg_Purchase_P,on="Gender")
DF2 = DF1.merge(Price_count, on="Gender")
Purchasing_Analysis_G = DF2.merge (ATP_PP, on="Gender")
Purchasing_Analysis_G
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 200]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_group = pd.cut(unique_SN["Age"], bins, labels=group_labels)
age_group
```

```python
unique_SN["Age Group"] = pd.cut(unique_SN["Age"], bins, labels=group_labels)
unique_SN.head()
```

```python
# Create a GroupBy object based upon "View Group"
unique_SN = unique_SN.groupby("Age Group")

# Find how many rows fall into each bin
age_group_count = pd.DataFrame(unique_SN["SN"].count())
age_group_count = age_group_count.rename(columns={"SN":"Total Count"})
age_group_count

# Get the average of each column within the GroupBy object
```

```python

age_group_percent = pd.DataFrame(age_group_count/players_count)
age_group_percent = age_group_percent.rename(columns={"Total Count":"Percentage of Players"})
age_group_percent["Percentage of Players"]=age_group_percent["Percentage of Players"].astype(float).map("{:,.2%}".format)
age_group_percent
```

```python
Age_Demographics = age_group_count.merge(age_group_percent, on = "Age Group")
Age_Demographics
```
